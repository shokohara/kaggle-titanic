```python
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
import csv
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline 
```

```python
def func(ldf):
#     arr = ["Name", "Age", "Sex", "SibSp", "Parch", "Ticket", "Fare", "Cabin", "Embarked"]
    ldf["Sex"] = ldf["Sex"].map({"female": 0, "male": 1}).astype(int)
    # 欠損値が存在するため失敗する
    # ldf["AltEmbarked"] = ldf["Embarked"].map({"C": 0, "Q": 1, "S": 2}).astype(int)
    ldf["Age"] = ldf["Age"].fillna(ldf.Age.median())
    ldf["Fare"] = ldf["Fare"].fillna(ldf.Fare.median())
    ldf["FamilySize"] = ldf["SibSp"] + ldf["Parch"] + 1
    # ldf = pd.concat([ldf, name_classifier(ldf.Name)], axis=1)
    return ldf.drop([], axis=1)
```

```python
df = pd.read_csv("train.csv")
df = func(df)
plt.figure(figsize=(10, 8)) #heatmap size
sns.heatmap(df.corr(), annot=True, cmap='plasma', linewidths=.5)
```
